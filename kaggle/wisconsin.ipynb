```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/data.csv')
df.info()
# df.isnull().sum()
```

```python
df.head()
```

```python
#데이터 전처리 
#문자-> 숫자 / 필요없는 칼럼 drop
df.diagnosis.unique()
```

```python
def function(row):
    if 'M' == row :
        return 1
    else :
        return 0
df['result'] = df.diagnosis.apply(function)
df.head()
```

```python
df1 = df.drop(columns = ['id','diagnosis','Unnamed: 32'])
df1.head()
```

```python
x = df1.drop(['result'],axis=1)
y = df1.result
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=2014,stratify=y)
#stratify => class 비율을 train/val 유지 (한 쪽에 쏠려서 분배되는 것을 방지)

print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score


model = RandomForestClassifier(n_estimators=50)
model.fit(x_train,y_train)
predict = model.predict(x_test)
print(f'정확도 : {accuracy_score(y_test, predict)}')
print(model.score(x_test,y_test))
"""
model.fit(x_train, y_train)
predict = model.predict(x_test)
submission = pd.DataFrame({
    'Num': df.Num,
    'Result' : predict
})
submission.to_csv('data/submission.csv',index=False)
"""
```
